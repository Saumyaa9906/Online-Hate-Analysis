```python
import numpy as np
import pandas as pd
import json
```

```python
users1 = pd.read_csv('snsusers_users.csv')
#users1.head()
print(users1.shape)
```

```python
users2 = pd.read_csv('snshashtags_users.csv')
#users2.head()
users2.shape
```

```python
#merge users
users = pd.concat([users1,users2])
users.head()
```

```python
users.shape
```

```python
#drop duplicates, clean
users = pd.DataFrame(users)
users.drop_duplicates(subset=['userId'], inplace=True)
users.head()
users.shape
```

```python
#users.to_csv('dataset_users.csv',index = False)
```

```python
tweets1 = pd.read_csv('snsusers_tweets.csv')
print(tweets1.shape)
tweets2 = pd.read_csv('snshashtags_tweets.csv')
print(tweets2.shape)
```

```python
#merge tweet data
tweets = pd.concat([tweets1,tweets2])
print(tweets.shape)

#drop duplicates, clean
tweets = pd.DataFrame(tweets)
tweets.drop_duplicates(subset=['tweetId'], inplace=True)
print(tweets.shape)
```

```python
tweets.head()
```

```python
#tweets.to_csv('dataset_tweets.csv',index = False)
```
