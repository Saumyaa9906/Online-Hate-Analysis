```python
import os
import pandas as pd
import numpy as np
import glob
```

```python
from pandas.io.json import json_normalize
import warnings

warnings.filterwarnings("ignore")
```

```python
flag = 1
for file in glob.glob("*.json"):
    df = pd.read_json(file, lines = True)
    if flag == 1:
        cols = df.columns
        df_users = pd.DataFrame(columns = cols)
        flag = 0
    else:
        df_users = pd.concat([df_users,df])
```

```python
#normalize user object
users = json_normalize(df_users['user'])
users.drop(['description', 'linkTcourl'], axis=1, inplace=True)
users.rename(columns={'id':'userId', 'url':'profileUrl'}, inplace=True)
users.head(2)
```

```python
#drop duplicates
users = pd.DataFrame(users)
users.drop_duplicates(subset=['userId'], inplace=True)
users.head(5)
```

```python
users.verified.unique()
```

```python
user_cols = ['username', 'displayname', 'userId','rawDescription','verified','created','followersCount',
       'friendsCount', 'statusesCount', 'favouritesCount', 'listedCount',
       'mediaCount', 'location']
```

```python
tweet_users = users[user_cols]
tweet_users.head(5)
```

```python
tweet_users.shape
```

```python
tweet_users.to_csv('snsusers_users.csv')
```

```python
#add userid column
user_id = []
for user in df_users['user']:
    uid = user['id']
    user_id.append(uid)
df_users['userId'] = user_id
```

```python
#clean
cols = ['date', 'renderedContent', 'id', 'userId', 'replyCount', 'retweetCount', 'likeCount', 'quoteCount', 'media', 'quotedTweet', 'mentionedUsers', 'lang']
tweets = df_users[cols]
tweets.rename(columns={'id':'tweetId'}, inplace=True)
tweets.head(5)
```

```python
#convert to dataframe, drop duplicates, keep only english tweets
tweets = pd.DataFrame(tweets)
tweets.drop_duplicates(subset=['tweetId'], inplace=True)
print(tweets.shape)
```
